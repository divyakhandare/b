<h3>Implement Gradient Descent Algorithm to find the local minima of a function. For example, find the local minima of the function y = (x+3)Â² starting from the point x = 2 </h3>

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def func(x):
    return (x+3)**2

def derivative(x):
    return 2*(x+3)
```

```python
gd = []
y_values = []
def gradient_descent(starting_x, learning_rate, num_iterations):
    x = starting_x
    gd.append(starting_x)
    y_values.append(func(starting_x))
    for i in range(num_iterations):
        grad = derivative(x)
        x = x - learning_rate * grad
        gd.append(x)
        y_values.append(func(x))
        print(f"Iteration {i+1}: x = {x} | f(x) ={func(x)}")
       
    return x
```

```python
starting_x = 2
learning_rate = 0.1
num_iterations = 100
min_x = gradient_descent(starting_x, learning_rate, num_iterations)
print(f"Local minima occurs at x = {min_x}")
```

```python
plt.figure(figsize = (8,6))
plt.scatter(gd,y_values)
plt.show()
```
